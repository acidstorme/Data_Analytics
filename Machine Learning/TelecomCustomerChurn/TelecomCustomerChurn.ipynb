```python
X = df_encoded.drop('Churn',axis=1)
Y = df_encoded['Churn']

## Split the data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.30)
print(X_train.shape,X_test.shape)
print(Y_train.shape,Y_test.shape)

## Scaling the data using standard scalar
ss = StandardScaler()

X_train.iloc[:,:] = ss.fit_transform(X_train.iloc[:,:])
X_test.iloc[:,:] = ss.transform(X_test.iloc[:,:])
```

# Logistic Regression

```python
log_reg = LogisticRegression()
log_reg.fit(X_train,Y_train)
```

```python
y_pred_LR = log_reg.predict(X_test)
```

```python
accuracy_score(Y_test,y_pred_LR)
```

```python
## Confusion matrix
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(log_reg, X_test, Y_test)  
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred_LR))
```

```python
from sklearn.metrics import roc_curve
from sklearn import metrics
fpr, tpr, thresholds = roc_curve(Y_test, y_pred_LR)

# plot the ROC curve
plt.plot(fpr, tpr)

# set limits for x and y axes
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])

# plot the straight line showing worst prediction for the model
plt.plot([0, 1], [0, 1],'r--')

# add plot and axes labels
# set text size using 'fontsize'
plt.title('ROC curve for Churn Classifier (Logistic Regression Model)', fontsize = 15)
plt.xlabel('False positive rate (1-Specificity)', fontsize = 15)
plt.ylabel('True positive rate (Sensitivity)', fontsize = 15)

# add the AUC score to the plot
# 'x' and 'y' gives position of the text
# 's' is the text 
# use round() to round-off the AUC score upto 4 digits
plt.text(x = 0.02, y = 0.9, s = ('AUC Score:', round(metrics.roc_auc_score(Y_test, y_pred_LR),4)))
                               
# plot the grid
plt.grid(True)
```

### Choosing best value for k

```python
error_rate = []

for i in range(1,30):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,Y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != Y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,30),error_rate,color='blue', linestyle='dashed', marker='X',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.show()
```

```python
knn_24 = KNeighborsClassifier(n_neighbors=24)
knn_24.fit(X_train,Y_train)

## y predicted
y_pred_knn_24 = knn_24.predict(X_test)

## accuracy
print('accuracy (k=24):',accuracy_score(Y_test,y_pred_knn_24))
```

```python
## Confusion matrix
plot_confusion_matrix(knn_24, X_test, Y_test)  
plt.show()
```

#### k = 24 has a 3% less error rate than deafult k=5, and hence gives a better result, so we can use this value.

```python
print(classification_report(Y_test,y_pred_knn_24))
```

```python
from sklearn.metrics import roc_curve
from sklearn import metrics
fpr, tpr, thresholds = roc_curve(Y_test, y_pred_knn_24)

# plot the ROC curve
plt.plot(fpr, tpr)

# set limits for x and y axes
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])

# plot the straight line showing worst prediction for the model
plt.plot([0, 1], [0, 1],'r--')

# add plot and axes labels
# set text size using 'fontsize'
plt.title('ROC curve for Churn Classifier (KNN 24 Model)', fontsize = 15)
plt.xlabel('False positive rate (1-Specificity)', fontsize = 15)
plt.ylabel('True positive rate (Sensitivity)', fontsize = 15)

# add the AUC score to the plot
# 'x' and 'y' gives position of the text
# 's' is the text 
# use round() to round-off the AUC score upto 4 digits
plt.text(x = 0.02, y = 0.9, s = ('AUC Score:', round(metrics.roc_auc_score(Y_test, y_pred_knn_24),4)))
                               
# plot the grid
plt.grid(True)
```

### Accuracy:

From above Precision scores, we understand that the Naive Bayes model has very high accuracy for predicting customers that will not churn but very low for those that do churn. On the other hand LR and KNN are very close with their precisions but LR is a more precise model for both values

### Recall (0,1):
#### Logistic Regression : 91%, 53%
#### Naive Bayes : 64%, 86%
#### KNN (k=24) : 89%, 51%
From above recall scores, we can see that Naive Bayes model can identify churned customers effectively but has very low accuracy in recalling those that do not churn. LR and KNN are again close with recall, with LR being a better model (by 2%) in both classes

### F1 scores (0,1):
#### Logistic Regression : 0.88, 0.59
#### Naive Bayes : 0.76, 0.59
#### KNN (k=24) : 0.86, 0.56
From above F1 scores, we can see that LR is the best model that can more accurately predict correct values

### AUC scores:
#### Logistic Regression : 0.72
#### Naive Bayes : 0.75
#### KNN (k=24) : 0.7
From above AUC scores, we can observe that there is not much difference between them, and since they are above 0.7 that means they are good predictions, though Naive Bayes has the highest AUC score.

### Comparing all the above data, we can affirm that:
#### Since AUC score of Naive Bayes is higher but it is not performing for other metrics, we can discard this model for our final selection. 
#### AUC score of KNN and LR is pretty much the same, hence we can use the other metrics to select a better model. Considering Accuracy, Precision, Recall and F1 scores, Logistic Regression model has performed better than K Nearest Neighbours model.

# Hence, from all of the above comparisons, we can finally select Logistic Regression Model as our best trained model.
