##### Replacing redundant part of string

```python
def get_unique(df, cols):
    for c in cols:
        print(df[c].value_counts())
        print("\n")
```

```python
cols = ["gender", "Partner", "Dependents", "PhoneService", "MultipleLines", 
        "InternetService", "OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", 
        "StreamingMovies", "Contract", "PaperlessBilling", "PaymentMethod", "Churn"]
get_unique(df,cols)
```

#### As can be seen from above data on number of unique categories in each non-continuous column, there are 6 columns with 2 categories and 10 columns with 3 categories.
### The target variable Churn shows that majorly people continue with their subscriptions and hence the data seems to be skewed towards Not Churning

```python
def encode_vars(df, cols):
    for col in cols:
        df[col] = df[col].astype('category')            #changing column type to category to encode
        d = dict(enumerate(df[col].cat.categories))     #creating dictionary to map code with category and print it
        print(col)
        print(d)
        print('\n')
        df[col] = df[col].cat.codes                    #converting col to encoded column
```

```python
cols = ["gender", "Partner", "Dependents", "PhoneService", "MultipleLines", 
        "InternetService", "OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", 
        "StreamingMovies", "Contract", "PaperlessBilling", "PaymentMethod", "Churn"]
```

```python
df_encoded = df.copy()
encode_vars(df_encoded,cols)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### We can observe the following information from the plots above for numeric data:
##### The customers' genders are distributed equally i.e. both male and female customers occupy half the dataset
##### Very few of the customers are senior citizens
##### There are about as many customers who are married as there are ones that are not
##### Most of the customers dont have any dependents. In fact, the number of customers that have dependents is less than half of those that do have
##### Large number of customers (almost 1750) keep subscribed for 10 months i,.e. they are newer customers, then second highest group of customers(~1100) stay for almost 65-70 months, who are part of the loyal group. Rest of the customers are distributed almost equally (500-750 customers) for other brackets of tenure in between
##### Almost all customers avail the phone services, with around 700 of them not having any phone service
##### Most people opt for paperless billing but there is quite a number of customers that dont want the paperless billing service
##### The group of customers that is largest pay about 20-30 dollars each month, which drops quite low in the 30-40 dollar bracket. From there number of customers increases as the charges increase till the 80-90 dollar bracket, while post that the number of customers drastically decrease as we reach the 110-20 dollar bracket.
##### The total charges data is skewed towards the left, indicating that most of the customers pay a total charge under 1000 dollars, and as the total charge increases the lesser number of people are present.

```python
corr_matrix = df_encoded[["gender", "SeniorCitizen", "Partner", "Dependents", "tenure", "PhoneService", "MultipleLines", 
        "InternetService", "OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", 
        "StreamingMovies", "Contract", "PaperlessBilling", "PaymentMethod", "MonthlyCharges", "TotalCharges", "Churn"]].corr()
corr_matrix
```

```python
plt.figure(figsize=(18,8))
sns.heatmap(corr_matrix,annot=True,cmap='plasma',vmin=-1,vmax=1)
plt.show()
```

#### From the heatmap, we can observe that:
##### Tenure and Total charges are highly correlated. This means that the customers that are loyal and stay subscribed for longest have higher total charges, which is obvious.
##### Tenure is also very highly correlated with contract feature, indicating that the type of contract has an important role to play in the subscription length of a customer.
##### Gender has low correlation with all other feature hence there is nothing to analyse there
##### Senior citizenship has no correlation with most features, but low correlation with Monthly charges, Billing method, Churn and negatively corellated with dependents.
##### Partner feature is slightly correlated with Total Charges and Contract, and a bit more correlated with tenure and Dependents. This means that those customers with partners tend to subscribe for longer and also have dependents.
##### Dependents feature is slightly correlated with Contract
##### Phone service is correlated with Internet services i.e. quite a few people that opt for phone services also add internet services to their package
##### Multiple lines option is fairly correlated with Monthly and Total charges, i.e. the higher the charges more chance of customer having multiple lines. It is also slightly correlated with most internet services and bit more correlated with tenure of customer
##### Most internet services such as online security, online backup, device protection, tech support etc fairly well correlated with Total Charges, Contract and Tenure of customer
##### Contract is fairly well correlated to Total Charges,Tech Supportm and other online services
##### Billing preference is not correlated with many deatures except Monthly Charges, indicating some customers like to get Paperless bills if they are paying higher charges monthly
#### Now looking at the Target variable column Churn, we can infer the following:
##### Churn is negatively correlated with Online Security, Tech Support, Tenure and Contract, i.e. those that do not churn have longer contracts and tenure and dont have tech support and online security.

```python
for i in df:
    if i == 'tenure':
        fig, ax = plt.subplots(figsize=(30, 10))
    elif i in ['MonthlyCharges', 'TotalCharges']:
        continue
    else:
        fig, ax = plt.subplots(figsize=(10, 5))
    
    ax = sns.countplot(x=i,hue='Churn', data=df)
    
    for c in ax.containers:
    
        # custom label calculates percent and add an empty string so 0 value bars don't have a number
        labels = [f'{h/df.shape[0]*100:0.1f}%' if (h := v.get_height()) > 0 else '' for v in c]
        ax.bar_label(c, labels=labels, label_type='edge')
    plt.show()
```

### Analysis churn wrt to other columns
##### There is no significant relationship with gender as the amount of customers that churn are almost equal for both
##### There is very high correlation with age to churning as almost half the Senior Citizens drop their subscription
##### Those with partners tend to churn less than customers that dont.
##### Those with dependents have a less amount of churn than those who dont, where churn rate is higher
##### Customers churn rate decreases as the tenure increases, with most customers dropping services in a month of starting
##### There is no discernible relationship of churn customers wrt Phone Service and Multiple Lines
##### More people who opted for Fiber Optic tend to churn than those that use DSL or no internet
##### For other Internet Services, more people tend to churn that do not opt for such services if they have internet
##### There is no difference in churn rate for customers who opt to stream movies and TV
##### Churn is highly correlated with Contract, and those that opt for monthly basis contracts tend to churn more than those that have year long contracts
##### Those who opt for paperless billing also churn more than those who dont
##### Customers that pay by electronic check have a very high rate of churn than those that pay by bank transfer or credit card
##### The customers that are still subscribing is 73.4%, and those that drop services are 26.6%, hence the data is skewed.
