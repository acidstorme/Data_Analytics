#### 13.Perform Correlation Analysis and write your key inferences. (Hint:- visualise using an appropriate plot) (1 point)

```python
corr_matrix = df[[" Income ", "Recency", "MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntGoldProds", "NumDealsPurchases", "NumWebPurchases", "NumCatalogPurchases", "NumStorePurchases", "NumWebVisitsMonth"]].corr() 
corr_matrix
```

From the heatmap it can be inferred that there is moderate corellation between most features. 

As income increases the amount spent on different products also increases slightly. Number of purchases made with deals has no correlation but catalog and store purchases increase slightly as income increases.

There is no corellation with Recency.

For amount spent on products there is higher corellation with Catalog purchases than Store purchases, though Web purchases is the least.

There is no corellation of deal purchases and any other feature except web visits.

Number of web visits per month is inversely corellated with Income. They also decrease slightly as more products are bought. As web visits per month increase so do purchases made with deals increase slightly, but purchases made on web, catalog and stores decreases as well.
